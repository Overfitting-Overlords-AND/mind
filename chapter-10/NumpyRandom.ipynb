```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['lines.markersize'] = 2
```

```python
xy = np.linspace([0., 0.], [5., 10.], 5)

fig, axs = plt.subplots()
axs.set_xlim(0, 100)
axs.set_ylim(0, 50)

axs.scatter(xy[:, 0:1], xy[:, 1:]);
```

```python
import matplotlib.pyplot as plt


def on_close(event):
    print('Closed Figure!')

fig = plt.figure()
fig.canvas.mpl_connect('close_event', on_close)

plt.text(0.35, 0.5, 'Close Me!', dict(size=30))
plt.show()
```
